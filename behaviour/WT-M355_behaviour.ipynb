Look at mean and variability within one session:

```python
# Code to look at the mean and std of velocity within one session:

fpath = os.path.join(data_dir, "WT-M355_session4.mat")  # change to the session you want
# load TrialData .mat file into a DataFrame
df1 = pyaldata.mat2dataframe(fpath, shift_idx_fields=True)

df1 = pyaldata.select_trials(df1, "success == 1")

df1 = pyaldata.restrict_to_interval(df1, "idx_mo", end_point_name=None, rel_start=-50, rel_end=700)

avg_vel1 = np.mean(df1.vel)

std_vel1 = np.std([trial_vel for trial_vel in df1.vel], axis = 0)

plt.plot(range(0, len(avg_vel1)*2, 2), avg_vel1, color='b')

plt.fill_between(range(0, len(avg_vel1)*2, 2), avg_vel1-std_vel1/2, avg_vel1+std_vel1/2, alpha=0.3, color='b')

plt.axvline(x=100, linestyle='--', color='r')

plt.legend(("mean (trial-averaged)",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("Velocity magnitude", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
```

Look at the mean and variability between sessions within one animal model:

```python
success_rate_list = []
avg_vel_list = []
avg_ga_list = []
avg_da_list = []
avg_xhand_list = []
avg_yhand_list = []
avg_zhand_list = []

for file_name in files:
    # Code to look at the mean and std of velocity within one session:

    fpath = os.path.join(data_dir, file_name)  # change to the session you want
    # load TrialData .mat file into a DataFrame
    df = pyaldata.mat2dataframe(fpath, shift_idx_fields=True)

    total_no_trials = len(df)

    df = pyaldata.select_trials(df, "success == 1")

    no_successful_trials = len(df)

    success_rate = no_successful_trials/total_no_trials

    success_rate_list.append(success_rate)

    if len(df) != 0:  # only consider sessions with successful trials

        df = pyaldata.restrict_to_interval(df, "idx_mo", end_point_name=None, rel_start=-50, rel_end=700)

        avg_vel = np.mean(df.vel)
        avg_vel_list.append(avg_vel)

        avg_ga = np.mean(df.ga)
        avg_ga_list.append(avg_ga)

        avg_da = np.mean(df.da)
        avg_da_list.append(avg_da)

        avg_xhand = np.mean([trial_hand[:,0] for trial_hand in df.hand], axis=0)
        avg_xhand_list.append(avg_xhand)

        avg_yhand = np.mean([trial_hand[:,1] for trial_hand in df.hand], axis=0)
        avg_yhand_list.append(avg_yhand)

        avg_zhand = np.mean([trial_hand[:,2] for trial_hand in df.hand], axis=0)
        avg_zhand_list.append(avg_zhand)



model_avg_vel = np.mean(avg_vel_list, axis=0)
model_std_vel = np.std(avg_vel_list, axis=0)

model_avg_ga = np.mean(avg_ga_list, axis=0)
model_std_ga = np.std(avg_ga_list, axis=0)

model_avg_da = np.mean(avg_da_list, axis=0)
model_std_da = np.std(avg_da_list, axis=0)

model_avg_xhand = np.mean(avg_xhand_list, axis=0)
model_std_xhand = np.std(avg_xhand_list, axis=0)

model_avg_yhand = np.mean(avg_yhand_list, axis=0)
model_std_yhand = np.std(avg_yhand_list, axis=0)

model_avg_zhand = np.mean(avg_zhand_list, axis=0)
model_std_zhand = np.std(avg_zhand_list, axis=0)
```

```python
plt.figure()
plt.plot(range(0, len(model_avg_vel)*2, 2), model_avg_vel, color='b')
plt.fill_between(range(0, len(model_std_vel)*2, 2), model_avg_vel-model_std_vel/2, model_avg_vel+model_std_vel/2, alpha=0.3, color='b')
plt.axvline(x=100, linestyle='--', color='r')
plt.legend(("mean",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("Velocity magnitude (m/s)", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title("Hand velocity magnitude across sessions for WT-M355", pad=10, fontsize=10)


plt.figure()
plt.plot(range(0, len(model_avg_ga)*2, 2), model_avg_ga, color='b')
plt.fill_between(range(0, len(model_std_ga)*2, 2), model_avg_ga-model_std_ga/2, model_avg_ga+model_std_ga/2, alpha=0.3, color='b')
plt.axvline(x=100, linestyle='--', color='r')
plt.legend(("mean",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("Grip aperture", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title("Grip aperture across sessions for WT-M355", pad=10, fontsize=10)


plt.figure()
plt.plot(range(0, len(model_avg_da)*2, 2), model_avg_da, color='b')
plt.fill_between(range(0, len(model_std_da)*2, 2), model_avg_da-model_std_da/2, model_avg_da+model_std_da/2, alpha=0.3, color='b')
plt.axvline(x=100, linestyle='--', color='r')
plt.legend(("mean",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("Digit angle", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title("Digit angle across sessions for WT-M355", pad=10, fontsize=10)


plt.figure()
plt.plot(range(0, len(model_avg_xhand)*2, 2), model_avg_xhand, color='b')
plt.fill_between(range(0, len(model_std_xhand)*2, 2), model_avg_xhand-model_std_xhand/2, model_avg_xhand+model_std_xhand/2, alpha=0.3, color='b')
plt.axvline(x=100, linestyle='--', color='r')
plt.legend(("mean",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("x-position (cm)", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title("x-component of hand trajectory across sessions for WT-M355", pad=10, fontsize=10)


plt.figure()
plt.plot(range(0, len(model_avg_yhand)*2, 2), model_avg_yhand, color='b')
plt.fill_between(range(0, len(model_std_yhand)*2, 2), model_avg_yhand-model_std_yhand/2, model_avg_yhand+model_std_yhand/2, alpha=0.3, color='b')
plt.axvline(x=100, linestyle='--', color='r')
plt.legend(("mean",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("y-position (cm)", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title("y-component of hand trajectory across sessions for WT-M355", pad=10, fontsize=10)


plt.figure()
plt.plot(range(0, len(model_avg_zhand)*2, 2), model_avg_zhand, color='b')
plt.fill_between(range(0, len(model_std_zhand)*2, 2), model_avg_zhand-model_std_zhand/2, model_avg_zhand+model_std_zhand/2, alpha=0.3, color='b')
plt.axvline(x=100, linestyle='--', color='r')
plt.legend(("mean",'std','movement onset'), fontsize=8)
plt.xlabel("Time(milliseconds)", fontsize=8)
plt.ylabel("z-position (cm)", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title("z-component of hand trajectory across sessions for WT-M355", pad=10, fontsize=10)
```
