```python
files_WTM350 = os.listdir(data_dir_WTM350)
fpath = os.path.join(data_dir_WTM350, files_WTM350[7])
df = pyaldata.mat2dataframe(fpath, shift_idx_fields=True)
df_successful = pyaldata.select_trials(df, "success == 1")  # only consider successful trials

fig = plt.figure(figsize = (5,5))
ax = plt.axes(projection='3d')
ax.grid()

for i in range(len(df_successful)):
    ax.plot(df_successful.hand[i][int(df_successful.idx_mo[i]):int(df_successful.idx_mend[i]),0], df_successful.hand[i][int(df_successful.idx_mo[i]):int(df_successful.idx_mend[i]),1], df_successful.hand[i][int(df_successful.idx_mo[i]):int(df_successful.idx_mend[i]),2])

ax.set_title('3D Scatter Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)
```
