```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example data: replace with your actual 3D coordinates
# Assuming each trial has n_points and is stored as an array of shape (n_points, 3)
# trials is a list of such arrays
trials = [
    np.random.rand(100, 3) * i for i in range(1, 6)
]

# Set up the figure and 3D axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot each trial
for i, trial in enumerate(trials):
    ax.plot(trial[:, 0], trial[:, 1], trial[:, 2], label=f'Trial {i+1}', linewidth=2)

# Customize the plot
ax.set_title('3D Hand Trajectories of Reaching Task', fontsize=20, fontweight='bold')
ax.set_xlabel('X axis', fontsize=15)
ax.set_ylabel('Y axis', fontsize=15)
ax.set_zlabel('Z axis', fontsize=15)

# Set ticks
ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis='both', which='minor', labelsize=10)

# Add a grid
ax.grid(True)

# Add legend
ax.legend(fontsize=12, loc='upper left', frameon=True)

# Set aspect ratio for better visualization
ax.set_box_aspect([1, 1, 1])  # Aspect ratio is 1:1:1

# Set view angle for better visualization
ax.view_init(elev=30, azim=120)

# Show the plot
plt.tight_layout()
plt.show()
```
